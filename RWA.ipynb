```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
```

```python
url = 'https://coinmarketcap.com/view/real-world-assets/'
r = requests.get(url)
```

```python
print(r)
```

```python
Soup = BeautifulSoup(r.text, 'lxml')
Soup
```

```python
tables = Soup.find_all('table', class_ = 'sc-14cb040a-3 dsflYb cmc-table')
print(tables)
```

```python
header = Soup.find_all('table')
header
```

```python
titles = []
for i in header:
    title = i.text
    titles.append(title)
print(titles)
```

```python
df = pd.DataFrame(columns=titles)
```

```python
Values = Soup.find_all('tr')
Values
```

```python
values = []

for i in Values:
    valoo = i.text
    values.append(valoo)
print(values)
```

```python
df
```

```python
df
```

```python
table = tables[0]

header = table.find('thead').find_all('th')
titles = [th.text.strip() for th in header]
print(titles)

values = []
rows = table.find('tbody').find_all('tr')
for row in rows:
    row_data = [td.text.strip() for td in row.find_all('td')]
    values.append(row_data)

# Create DataFrame
df = pd.DataFrame(values, columns=titles)
print(df)
```

```python
df = pd.DataFrame(values, columns=titles)

img_tags = Soup.find_all('img', class_='coin-logo')

links = [tag.get('src') for tag in img_tags]

df['Image Link'] = ''  # Initialize the new column with empty strings

# Iterate over the DataFrame index and update the 'Image Link' column
for idx in df.index:
    if idx < len(links):
        df.at[idx, 'Image Link'] = links[idx]

print(df)
```
