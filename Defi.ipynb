```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
import time
```

```python
url = 'https://coinmarketcap.com/view/defi/'
```

```python
r = requests.get(url)
r
```

```python
Soup = BeautifulSoup(r.text, 'lxml')
```

```python
table = Soup.find_all('table')
```

```python
headers = []

for i in table:
    table = i.text
    headers.append(table)
print(headers)
```

```python
header = []

for i in headers:
    head = i.text
    header.append(head)
print(header)
```

```python
values = Soup.find_all('tr')
```

```python
Values = []

for i in values:
    val = i.text
    Values.append(val)
print(Values)
```

```python
rows = Soup.find_all('tr')       
values = []

for row in rows:
    val = [cell.text for cell in row.find_all('td')]
    values.append(val)

# Creating DataFrame with extracted values
df = pd.DataFrame(values, columns=header)
print(df)
```

```python
img_tags = Soup.find_all('img', class_='coin-logo')

links = [tag.get('src') for tag in img_tags]

df['Image Link'] = ''  # Initialize the new column with empty strings

# Iterate over the DataFrame index and update the 'Image Link' column
for idx in df.index:
    if idx < len(links):
        df.at[idx, 'Image Link'] = links[idx]

print(df)
```
